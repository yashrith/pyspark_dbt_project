### **Spark Streaming**

```python
entities = ['customers', 'trips', 'locations', 'payments', 'vehicles', 'drivers']
```

```python
for entity in entities:
      df_batch = spark.read.format("csv")\
      .option("header", True)\
      .option("inferSchema", True)\
      .load(f"/Volumes/workspace/source_py_dbt/source_data/{entity}")  

      schema_entity = df_batch.schema

      df = (spark.readStream.format("csv")
      .option("header",True)
      .schema(schema_entity)
      .load(f"/Volumes/workspace/source_py_dbt/source_data/{entity}"))

      (df.writeStream.format("delta")
      .outputMode("append")
      .option("checkpointLocation", f"/Volumes/workspace/bronze/checkpoint/{entity}")
      .trigger(once=True) 
      .toTable(f"workspace.bronze.{entity}"))
```
