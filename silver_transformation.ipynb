## **CUSTOMERS**

```python
df_cust = spark.read.table("workspace.bronze.customers")
```

```python
display(df_cust)
```

```python
df_cust = df_cust.withColumn("domain", split(col("email"), "@")[1])
display(df_cust)
```

```python
df_cust = df_cust.withColumn("phone_number", regexp_replace("phone_number",r"[^0-9]", ""))
display(df_cust)
```

```python
df_cust = df_cust.withColumn("full_name", concat_ws(" ", col("first_name"), col("last_name")))
df_cust = df_cust.drop("first_name", "last_name")
display(df_cust)
```

```python
from typing import List
from pyspark.sql import DataFrame, SparkSession
from pyspark.sql.window import Window
from pyspark.sql.functions import concat_ws, col, row_number, desc, current_timestamp
from delta.tables import DeltaTable

class Transformations:
    
    def __init__(self, spark: SparkSession):
        # We store the spark session to use it in the upsert method
        self.spark = spark

    def dedup(self, df: DataFrame, dedup_cols: List[str], cdc: str) -> DataFrame:
        # 1. Use concat_ws to create a unique key (handles NULLs better than concat)
        df = df.withColumn("dedupKey", concat_ws("-", *[col(c) for c in dedup_cols]))
        
        # 2. Use .desc() so the NEWEST record gets row_number 1
        window_spec = Window.partitionBy("dedupKey").orderBy(col(cdc).desc())
        
        df = df.withColumn("dedupCounts", row_number().over(window_spec))
        
        # 3. Filter to keep only the latest record and drop helper columns
        df = df.where(col("dedupCounts") == 1).drop("dedupKey", "dedupCounts")
        
        return df

    def process_timestamp(self, df: DataFrame) -> DataFrame:
        # Adds a column showing exactly when the processing ran
        return df.withColumn("process_timestamp", current_timestamp())

    def upsert(self, df: DataFrame, key_cols: List[str], table_name: str, cdc_col: str):
        # 1. Dynamically build the join condition: "src.id = trg.id AND ..."
        merge_condition = " AND ".join([f"src.{i} = trg.{i}" for i in key_cols])
        
        # 2. Access the Delta Table via its catalog name
        dlt_obj = DeltaTable.forName(self.spark, f"workspace.silver.{table_name}")
        
        # 3. Execute Merge: 
        # Update existing rows ONLY if source data is newer (src.cdc >= trg.cdc)
        (dlt_obj.alias("trg")
         .merge(df.alias("src"), merge_condition)
         .whenMatchedUpdateAll(condition = f"src.{cdc_col} >= trg.{cdc_col}")
         .whenNotMatchedInsertAll()
         .execute())
        
        return 1
```

```python
# Fix: Pass the 'spark' session to the constructor
cust_obj = Transformations(spark) 

# Fix: The dedup method now handles the logic internally
cust_df_trns = cust_obj.dedup(df_cust, ['customer_id'], 'last_updated_timestamp')
df_cust = cust_obj.process_timestamp(cust_df_trns)
display(df_cust)
```

```python
if not spark.catalog.tableExists("workspace.silver.customers"):

    (df_cust.write.format("delta").mode("append").saveAsTable("workspace.silver.customers"))

else:

    cust_obj.upsert(df_cust,['customer_id'], 'customers', 'last_updated_timestamp')
```

### **PAYMENTS**

```python
df_pay = spark.read.table("workspace.bronze.payments")
display(df_pay)
```

```python
df_pay = (df_pay.withColumn("online_payment_status",
                            when(((col('payment_method')=='Card') & (col('payment_status') =='Success')), "online-success")
                            .when(((col('payment_method')=='Card') & (col('payment_status') =='Failed')), "online-failed")
                            .when(((col('payment_method')=='Card') & (col('payment_status') =='Pending')), "online-pending")
                            
                            .when(((col('payment_method')=='Cash') & (col('payment_status') =='Success')), "cash-success")
                            .when(((col('payment_method')=='Cash') & (col('payment_status') =='Failed')), "cash-failed")
                            .when(((col('payment_method')=='Cash') & (col('payment_status') =='Pending')), "cash-pending")

                            .when(((col('payment_method')=='Wallet') & (col('payment_status') =='Success')), "wallet-success")
                            .when(((col('payment_method')=='Wallet') & (col('payment_status') =='Failed')), "wallet-failed")
                            .when(((col('payment_method')=='Wallet') & (col('payment_status') =='Pending')), "wallet-pending")
                            ))
display(df_pay)
```

```python
# Fix: Pass the 'spark' session to the constructor
pay_obj = Transformations(spark) 

# Fix: The dedup method now handles the logic internally
df_pay_trns = pay_obj.dedup(df_pay, ['payment_id'], 'last_updated_timestamp')
df_pay = location_obj.process_timestamp(df_pay_trns)
display(df_pay)
```

```python
if not spark.catalog.tableExists("workspace.silver.payments"):

    (df_pay.write.format("delta").mode("append").saveAsTable("workspace.silver.payments"))

else:

    pay_obj.upsert(df_pay,['payment_id'], 'payments', 'last_updated_timestamp')
```
